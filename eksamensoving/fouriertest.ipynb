```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, ifft, fftshift

def f(x):
    return np.sin(2*np.pi/L*2*x) + np.cos(2*np.pi/L*5*x)

L = np.pi*4
N = 20
x = np.linspace(0, L, 100)

plt.plot(x, f(x))
plt.show()
```

```python
xSampling = np.linspace(0, L, N, endpoint = False)
fcs = f(xSampling)
fcsHat = fft(fcs)

plt.stem(fcsHat.real)
plt.show()

plt.stem(fcsHat.imag)
plt.show()
```

```python
fcsHatShift = fftshift(fcsHat)

plt.stem(fcsHatShift.real)
plt.show()
plt.stem(fcsHatShift)
plt.show()
```

```python
xFreqs = fftfreq(N, d=1/N)
xFreqsShifted = fftshift(xFreqs)

plt.stem(xFreqsShifted, fcsHatShift.real)
plt.show()
plt.stem(xFreqsShifted, fcsHatShift.imag)
plt.show()
```

```python
def g(x):
    return 0.5*np.sin((2+20)*np.pi/L*2*x) + f(x)

xcs = np.linspace(0, L, N, endpoint = False)
gcs = g(xcs)
gcsHat = fft(gcs)
gcsHatShifted = fftshift(gcsHat)
xFreqs = fftfreq(N, d=1/N)
xFreqsShifted = fftshift(xFreqs)

plt.stem(xFreqsShifted, gcsHatShifted.real)
plt.show()

plt.stem(xFreqsShifted, gcsHatShifted.imag)
plt.show()
```
